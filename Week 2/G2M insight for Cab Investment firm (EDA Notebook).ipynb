### Hypothesis No 1: Which cab company, has the biggest profit ?

I create a new column, that has the proft for every transaction

```python
cab_data_df["Profit"] = cab_data_df["Price Charged"] - cab_data_df["Cost of Trip"]
cab_data_df
```

Then, I create two new data frames for each cab company, one for the Pink cab and another one for the Yellow cab and I calculate the mean and median of the column Profit, in order to understand which one is more profitable

```python
company_groups = cab_data_df.groupby('Company')

pink_cab_data_df = company_groups.get_group('Pink Cab')
yellow_cab_data_df = company_groups.get_group('Yellow Cab')
```

```python
pink_cab_data_df
```

```python
yellow_cab_data_df
```

```python
mean_profit_pink_cab = pink_cab_data_df["Profit"].mean()
median_profit_pink_cab = pink_cab_data_df["Profit"].median()
print("Mean value profit for pink cab: ", mean_profit_pink_cab)
print("Median value profit for pink cab: ", median_profit_pink_cab)
```

```python
mean_profit_yellow_cab = yellow_cab_data_df["Profit"].mean()
median_profit_yellow_cab = yellow_cab_data_df["Profit"].median()
print("Mean value profit for yellow cab: ", mean_profit_yellow_cab)
print("Median value profit for yellow cab: ", median_profit_yellow_cab)
```

I visualize the results.

```python
pink_cab_values = [mean_profit_pink_cab, median_profit_pink_cab]
yellow_cab_values = [mean_profit_yellow_cab, median_profit_yellow_cab]
index = ['Mean Profit', 'Median Profit']
df = pd.DataFrame({'Pink Cab': pink_cab_values, 'Yellow Cab': yellow_cab_values}, index = index)
ax = df.plot.bar(color = ["pink","yellow"], rot = 0)
```

Overall, the yellow cab has much more profits than the pink cab.

### Hypothesis No 2: In which city, every cab company has bigger profit ?

```python
every_city_df = pd.merge(cab_data_df[["Company", "City", "Profit"]], city_data_df, on = ["City"], how = 'outer')
every_city_df
```

We drop all the rows where there is NaN, because SAN FRANCISCO CA does not use Pink or Yellow Cab

```python
every_city_df = every_city_df.dropna()
```

I calculate the mean profit for each cab company, for every city and create a new data frame that has this information.

```python
city_groups = every_city_df.groupby('City')

city_names = []
pink_cab_every_city_mean_profit_values = []
yellow_cab_every_city_mean_profit_values = []
for current_city, current_city_df in city_groups:
    
    current_cab_company_groups = current_city_df.groupby('Company')
    current_pink_cab_mean_profit_value = current_cab_company_groups.get_group('Pink Cab')["Profit"].mean()
    current_yellow_cab_mean_profit_value = current_cab_company_groups.get_group('Yellow Cab')["Profit"].mean()
    
    city_names.append(current_city)
    pink_cab_every_city_mean_profit_values.append(current_pink_cab_mean_profit_value)
    yellow_cab_every_city_mean_profit_values.append(current_yellow_cab_mean_profit_value)
    
every_city_mean_profit_dict = {
    "City": city_names,
    "Pink Cab Mean Profit": pink_cab_every_city_mean_profit_values,
    "Yellow Cab Mean Profit": yellow_cab_every_city_mean_profit_values
}

every_city_mean_profit_df = pd.DataFrame(every_city_mean_profit_dict, index = city_names)
every_city_mean_profit_df
```

I visualize the results in a bar plot

```python
ax = every_city_mean_profit_df.plot.bar(x = "City", title = "Mean Profit for Every City", color = ["pink","yellow"], rot = 0, figsize = (30, 10))
```

Conclusions:
- Except TUCSON AZ city, in every other city, yellow cab company has bigger profits than pink cab company
- So, most people prefer yellow cab rather than pink cab in almost every city, except TUCSON AZ
- The city with the bigger profit is New York City, for both companies

### Hypothesis No 4: For every company, which ages use a cab more often ?

```python
transaction_customer_df = pd.merge(customer_id_data_df[["Customer ID", "Age"]], transaction_id_data_df[["Transaction ID", "Customer ID"]], on = ["Customer ID"], how = 'outer')

transaction_customer_cab_df = pd.merge(cab_data_df[["Transaction ID", "Company", "Profit"]], transaction_customer_df, on = ["Transaction ID"], how = 'outer')
transaction_customer_cab_df
```

```python
transaction_customer_cab_df = transaction_customer_cab_df.dropna()
transaction_customer_cab_df
```

First we need to find out the min and the max value from the column Age.

```python
transaction_customer_cab_df["Age"].describe()
```

As we can see, we have ages from 18 years old to 65 years old. So, we are going to create a new column that is going to be age range.

```python
# create a list of our conditions
conditions = [
    (transaction_customer_cab_df['Age'] <= 18),
    (transaction_customer_cab_df['Age'] >= 20) & (transaction_customer_cab_df['Age'] <= 29),
    (transaction_customer_cab_df['Age'] >= 30) & (transaction_customer_cab_df['Age'] <= 39),
    (transaction_customer_cab_df['Age'] >= 40) & (transaction_customer_cab_df['Age'] <= 49),
    (transaction_customer_cab_df['Age'] >= 50) & (transaction_customer_cab_df['Age'] <= 59),
    (transaction_customer_cab_df['Age'] >= 60) & (transaction_customer_cab_df['Age'] <= 69),
    ]

# create a list of the values we want to assign for each condition
values = ["10's", "20's", "30's", "40's", "50's", "60's"]

# create a new column and use np.select to assign values to it using our lists as arguments
transaction_customer_cab_df['Age range'] = np.select(conditions, values)

# display updated DataFrame
transaction_customer_cab_df
```

```python
company_groups = transaction_customer_cab_df.groupby('Company')

company_names = []
tens_values = []
twenties_values = []
thirties_values = []
fourties_values = []
fifties_values = []
sixties_values = []
for current_company, current_company_df in company_groups:
    
    current_age_range_groups = current_company_df.groupby('Age range')
    current_tens_value = current_age_range_groups.get_group("10's").shape[0]
    current_twenties_value = current_age_range_groups.get_group("20's").shape[0]
    current_thirties_value = current_age_range_groups.get_group("30's").shape[0]
    current_fourties_value = current_age_range_groups.get_group("40's").shape[0]
    current_fifties_value = current_age_range_groups.get_group("50's").shape[0]
    current_sixties_value = current_age_range_groups.get_group("60's").shape[0]
    
    company_names.append(current_company)
    tens_values.append(current_tens_value)
    twenties_values.append(current_twenties_value)
    thirties_values.append(current_thirties_value)
    fourties_values.append(current_fourties_value)
    fifties_values.append(current_fifties_value)
    sixties_values.append(current_sixties_value)
    
age_range_dict = {
    "Company": company_names,
    "10's": tens_values,
    "20's": twenties_values,
    "30's": thirties_values,
    "40's": fourties_values,
    "50's": fifties_values,
    "60's": sixties_values
}

company_age_range_df = pd.DataFrame(age_range_dict, index = company_names)
company_age_range_df
```

```python
ax = company_age_range_df.plot.bar(x = "Company", rot = 0, figsize = (10, 10))
```

It is obvious that people between 20 and 29 use cab more often than the other ages in both cab companies.

### Hypothesis No 5: For every company, which season is considered more profitable?


```python
cab_data_df
```

```python
winter_list = [12, 1, 2]
spring_list = [3, 4, 5]
summer_list = [6, 7, 8]

# A function that takes as input a date and return the month of the date.
def find_season(a_date):
    temp_list = a_date.split("-")
    month = int(temp_list[1])
    if (month in winter_list):
        return "Winter"
    elif (month in spring_list):
        return "Spring"
    elif (month in summer_list):
        return "Summer"
    else:
        return "Autumn"



cab_data_df["Season"] = cab_data_df['Date of Travel'].apply(find_season)
```

```python
cab_data_df
```

```python
company_groups = cab_data_df.groupby('Company')

company_names = []
winter_mean_profit_values = []
spring_mean_profit_values = []
summer_mean_profit_values = []
autumn_mean_profit_values = []
for current_company, current_company_df in company_groups:
    
    current_season_groups = current_company_df.groupby('Season')
    current_winter_mean_profit_value = current_season_groups.get_group("Winter")["Profit"].mean()
    current_spring_mean_profit_value = current_season_groups.get_group("Spring")["Profit"].mean()
    current_summer_mean_profit_value = current_season_groups.get_group("Summer")["Profit"].mean()
    current_autumn_mean_profit_value = current_season_groups.get_group("Autumn")["Profit"].mean()
    
    company_names.append(current_company)
    winter_mean_profit_values.append(current_winter_mean_profit_value)
    spring_mean_profit_values.append(current_spring_mean_profit_value)
    summer_mean_profit_values.append(current_summer_mean_profit_value)
    autumn_mean_profit_values.append(current_autumn_mean_profit_value)
    
season_dict = {
    "Company": company_names,
    "Winter": winter_mean_profit_values,
    "Spring": spring_mean_profit_values,
    "Summer": summer_mean_profit_values,
    "Autumn": autumn_mean_profit_values,
}

season_df = pd.DataFrame(season_dict, index = company_names)
season_df
```

```python
ax = season_df.plot.bar(x = "Company", rot = 0, figsize = (10, 10))
```

We can point out that for the Pink company, winter is more profitable and for the Yellow company Spring is more profitable.
